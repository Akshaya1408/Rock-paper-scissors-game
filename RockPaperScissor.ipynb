```python
import random

player = 0
cpu = 0

print("Score 3 points to win the game!")

while player<3 and cpu<3:
  cpu_choice = random.choice(["rock","paper","scissors"])
  player_choice = input(" Enter your choice. Rock or Paper or Scissors ?! : ")

  print(f"CPU : {cpu_choice} vs PLAYER : {player_choice}")

  if player_choice.lower() == cpu_choice:
    print("Tie! No points:(")
  elif player_choice.lower() == "rock":
    if cpu_choice == "scissors":
      player += 1
      print("Wohoo! You scored one point!")
    elif cpu_choice == "paper":
      cpu += 1
      print("Computer scores one point!")

  elif player_choice.lower() == "scissors":
    if cpu_choice == "paper":
      player += 1
      print("Wohoo! You scored one point!")
    elif cpu_choice == "rock":
      cpu += 1
      print("Computer scores one point!")

  elif player_choice.lower() == "paper":
    if cpu_choice == "rock":
      player += 1
      print("Wohoo! You scored one point!")
    elif cpu_choice == "scissor":
      cpu += 1
      print("Computer scores one point!")

    else:
      print("Invalid input")
  
print("You win!!" if player > cpu else "Computer wins!!")
```
